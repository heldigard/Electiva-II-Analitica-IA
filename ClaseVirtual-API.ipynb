#### **1. Introducción a las APIs: Las Puertas de Acceso a los Datos del Mundo**

API (Interfaz de Programación de Aplicaciones). Una buena analogía es pensar en una API como un mesero en un restaurante:

*   **Usted (el cliente/analista):** Quiere obtener algo específico (un plato de comida/datos).
*   **La Cocina (el servidor/fuente de datos):** Tiene lo que usted necesita, pero no puede entrar directamente a la cocina a prepararlo.
*   **El Mesero (la API):** Toma su pedido (petición), lo lleva a la cocina y le trae de vuelta lo que solicitó (respuesta).

En el mundo digital, las APIs permiten que diferentes aplicaciones se comuniquen entre sí e intercambien datos de una manera estandarizada. Para un analista de datos, esto significa acceso a datos en tiempo real de diversas fuentes: redes sociales, mercados financieros, información gubernamental, datos meteorológicos y mucho más.

#### **2. Selección de APIs Gratuitas para Proyectos de Análisis**

No todas las APIs son iguales. Para empezar, es ideal utilizar APIs que no requieran autenticación compleja (como OAuth) y que ofrezcan datos interesantes para analizar. Aquí tiene una selección de APIs excelentes para principiantes:

| API | Descripción de los Datos | Requiere Registro | Ideal Para... |

| **JSONPlaceholder** | Proporciona datos falsos pero realistas (usuarios, posts, fotos, etc.) en formato JSON. | No | Aprender a hacer peticiones y procesar la estructura de los datos. |

| **OpenWeatherMap** | Ofrece datos meteorológicos actuales y pronósticos para ciudades de todo el mundo. | Sí (para obtener una API Key gratuita) | Trabajar con datos numéricos, series de tiempo y geolocalización. |
| **REST Countries** | Información detallada sobre países: población, área, capital, idiomas, monedas, etc. | No | Análisis de datos demográficos y geográficos. |
| **TheCocktailDB** | Una base de datos de recetas de cócteles con ingredientes, imágenes e instrucciones. | Sí (para obtener una API Key gratuita) | Análisis de texto y datos categóricos. |

Para esta clase, vamos a utilizar **REST Countries** por su simplicidad y la riqueza de sus datos.

#### **3. ¡Manos a la Obra! Consultando una API con Python en Google Colab**

Ahora, guíe a sus estudiantes a través de estos pasos en un nuevo notebook de Google Colab.



**Paso 1: Instalar la librería `requests`**

Aunque `requests` usualmente viene preinstalada en Google Colab, es una buena práctica asegurarse.



```python
# No es estrictamente necesario en Colab, pero es buena práctica en otros entornos.
!pip install requests
```

**Paso 2: Importar las librerías necesarias**

Necesitaremos `requests` para hacer la petición a la API y `pandas` para manejar los datos.

```python
import requests
import pandas as pd
```

**Paso 3: Realizar la petición a la API**

Vamos a pedirle a la API de REST Countries toda la información que tiene sobre todos los países.

```python
# La URL del 'endpoint' que nos dará la información de todos los países
url = "https://restcountries.com/v3.1/all?fields=name,flags,capital,currencies,languages,population,status,region"

# Utilizamos requests.get() para hacer la petición
response = requests.get(url)

# Es una buena práctica verificar el estado de la respuesta
# Un código 200 significa que todo salió bien.
print(f"Status Code: {response.status_code}")
```

**Paso 4: Convertir la respuesta a formato JSON**

La mayoría de las APIs modernas devuelven los datos en formato JSON (JavaScript Object Notation), que es muy fácil de trabajar en Python.

```python
# Convertimos la respuesta en un objeto JSON (una lista de diccionarios en este caso)
data = response.json()

# Veamos la información de los primeros dos países para entender la estructura
print(data[0:2])
```

**Paso 5: Cargar los datos en un DataFrame de Pandas**

¡Aquí es donde comienza la magia del análisis de datos! Un DataFrame nos proporciona una estructura tabular con la que es mucho más sencillo trabajar.

```python
# Creamos el DataFrame directamente desde la lista de diccionarios
df_paises = pd.DataFrame(data)

# Mostramos las primeras 5 filas del DataFrame
df_paises.head()
```

#### **4. Exploración y Análisis de Datos (EDA)**

Una vez que los datos están en un DataFrame, podemos empezar a hacer preguntas.

**Pregunta 1: ¿Cuáles son las columnas que tenemos y qué tipo de datos contienen?**



```python
# Obtenemos información general del DataFrame
df_paises.info()
```

*Observación:* Notarán que algunas columnas contienen objetos (listas o diccionarios), como 'name', 'currencies', 'languages'. ¡Esto es interesante y requerirá un poco de pre-procesamiento para análisis más profundos!



**Pregunta 2: ¿Cuáles son los 10 países más poblados del mundo?**

```python
# Ordenamos el DataFrame por la columna 'population' de mayor a menor y tomamos los 10 primeros
top_10_poblacion = df_paises.sort_values(by='population', ascending=False).head(32)

# Mostramos el resultado
print(top_10_poblacion[['name', 'population']])
```

*Observación:* La columna 'name' es un diccionario. Para mostrar solo el nombre común, tendríamos que acceder a él. Hagámoslo más limpio.

**Limpiando la columna de nombres:**

```python
# Extraemos el nombre común de la columna 'name' que es un diccionario
df_paises['common_name'] = df_paises['name'].apply(lambda x: x.get('common'))

# Ahora, repitamos la pregunta con el nombre limpio
top_10_poblacion = df_paises.sort_values(by='population', ascending=False).head(10)
print(top_10_poblacion[['common_name', 'population', 'region']])
```

**Pregunta 3: ¿Cuántos países hay por región (continente)?**

```python
# Contamos los valores de la columna 'region'
conteo_por_region = df_paises['region'].value_counts()

print(conteo_por_region)
```

**Visualización de los resultados**

Una parte crucial del análisis es la visualización. Usemos `matplotlib` para visualizar el conteo por región.


```python
import matplotlib.pyplot as plt

# Creamos un gráfico de barras
conteo_por_region.plot(kind='bar', figsize=(10, 6))
plt.title('Número de Países por Región')
plt.xlabel('Región')
plt.ylabel('Cantidad de Países')
plt.xticks(rotation=45)
plt.show()
```
