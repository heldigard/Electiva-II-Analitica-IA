## üß† Parte 2: El Arte del Prompting

### ¬øQu√© es un Prompt?

Un **prompt** es la instrucci√≥n o pregunta que le das al modelo de IA. La calidad de tu prompt determina la calidad de la respuesta.

### üîÑ Metodolog√≠a de Trabajo: El Ciclo de An√°lisis con IA

```
1Ô∏è‚É£ PLANIFICAR ‚Üí ¬øQu√© quiero analizar?
      ‚¨áÔ∏è
2Ô∏è‚É£ EJEMPLIFICAR ‚Üí Mostrar datos de ejemplo al LLM
      ‚¨áÔ∏è
3Ô∏è‚É£ GENERAR ‚Üí Pedir c√≥digo con prompts claros
      ‚¨áÔ∏è
4Ô∏è‚É£ EJECUTAR ‚Üí Correr el c√≥digo en Colab
      ‚¨áÔ∏è
5Ô∏è‚É£ INTERPRETAR ‚Üí ¬øTiene sentido el resultado?
      ‚¨áÔ∏è
6Ô∏è‚É£ ITERAR ‚Üí Refinar y repetir
```

### üìù Anatom√≠a de un Buen Prompt

Un prompt efectivo tiene estos componentes:

| Componente | Descripci√≥n | Ejemplo |
|------------|-------------|--------|
| **Contexto** | Describe la situaci√≥n | "Tengo un dataset de pel√≠culas de Netflix..." |
| **Datos de ejemplo** | Muestra la estructura | "Las columnas son: title, type, release_year..." |
| **Tarea espec√≠fica** | Qu√© quieres lograr | "Quiero saber cu√°ntas pel√≠culas hay por a√±o" |
| **Formato deseado** | C√≥mo quieres el resultado | "Dame el c√≥digo en Python usando pandas" |

### üéØ Ejemplo: Prompt Malo vs Prompt Bueno

#### ‚ùå Prompt MALO:
```
"Analiza mis datos"
```

**Problemas:** No hay contexto, no hay datos de ejemplo, no hay objetivo claro.

---

#### ‚úÖ Prompt BUENO:
```
Tengo un dataset de Netflix con las siguientes columnas:
- show_id: identificador √∫nico
- type: 'Movie' o 'TV Show'

- duration: duraci√≥n (ej: '90 min' o '2 Seasons')

Ejemplo de las primeras filas:
| show_id | type  | title      | release_year | duration |

| s2      | TV Show | Friends  | 1994         | 10 Seasons |
```

## üßπ Parte 5: Limpieza B√°sica de Datos

### 5.1 Identificar valores faltantes

Los valores faltantes (NaN) son muy comunes en datasets reales. Es importante identificarlos.

````python
# Contar valores faltantes por columna
valores_faltantes = df.isnull().sum()

# Mostrar solo las columnas con valores faltantes
print("üìä Columnas con valores faltantes:")
print("=" * 40)
print(valores_faltantes[valores_faltantes > 0].sort_values(ascending=False))
````

### 5.2 Visualizar valores faltantes

````python
# Crear gr√°fico de valores faltantes
plt.figure(figsize=(10, 5))

# Calcular porcentaje de valores faltantes
porcentaje_faltantes = (df.isnull().sum() / len(df) * 100).sort_values(ascending=True)

# Crear gr√°fico de barras horizontales
porcentaje_faltantes.plot(kind='barh', color='coral')

plt.title('üìä Porcentaje de Valores Faltantes por Columna', fontsize=14, fontweight='bold')
plt.xlabel('Porcentaje (%)')
plt.ylabel('Columna')

# Agregar l√≠nea vertical en 0%
plt.axvline(x=0, color='black', linewidth=0.5)

plt.tight_layout()
plt.show()
````

## üìà Parte 6: An√°lisis B√°sico con Prompting

### 6.1 An√°lisis por A√±o de Lanzamiento

Vamos a analizar c√≥mo se distribuyen los t√≠tulos por a√±o de lanzamiento.

````python
# Contar t√≠tulos por a√±o de lanzamiento
titulos_por_a√±o = df['release_year'].value_counts().sort_index()

# Filtrar √∫ltimos 20 a√±os para mejor visualizaci√≥n
titulos_recientes = titulos_por_a√±o[titulos_por_a√±o.index >= 2005]

# Crear gr√°fico
plt.figure(figsize=(12, 5))
titulos_recientes.plot(kind='bar', color='steelblue', edgecolor='white')

plt.title('üìä T√≠tulos de Netflix por A√±o de Lanzamiento (2005-2021)', fontsize=14, fontweight='bold')
plt.xlabel('A√±o de Lanzamiento')
plt.ylabel('Cantidad de T√≠tulos')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
````

### ü§î Interpretaci√≥n

**Preguntas para reflexionar:**
1. ¬øEn qu√© a√±o hay m√°s t√≠tulos?
2. ¬øVes alguna tendencia a lo largo del tiempo?
3. ¬øTiene sentido este patr√≥n? ¬øPor qu√©?

### 6.2 Comparaci√≥n: Pel√≠culas vs Series por A√±o

````python
# Agrupar por a√±o y tipo
por_a√±o_tipo = df[df['release_year'] >= 2005].groupby(['release_year', 'type']).size().unstack(fill_value=0)

# Crear gr√°fico
fig, ax = plt.subplots(figsize=(12, 5))

por_a√±o_tipo.plot(kind='bar', ax=ax, width=0.8)

plt.title('üìä Pel√≠culas vs Series por A√±o de Lanzamiento', fontsize=14, fontweight='bold')
plt.xlabel('A√±o de Lanzamiento')
plt.ylabel('Cantidad de T√≠tulos')
plt.legend(title='Tipo')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
````

## üéØ Parte 7: T√©cnicas Avanzadas de Prompting

### 7.1 El Prompt de "Contexto Completo"

Cuando necesites hacer an√°lisis m√°s complejos, usa este formato:

```
CONTEXTO:
[Describe la situaci√≥n y los datos]

DATOS:
[Muestra ejemplos espec√≠ficos de tus datos]

OBJETIVO:
[Qu√© quieres lograr]

RESTRICCIONES:
[Cualquier limitaci√≥n o preferencia]

FORMATO DE SALIDA:
[C√≥mo quieres el c√≥digo o resultado]
```

### 7.2 Ejemplo de Prompt Completo:

```
CONTEXTO:
Soy analista de datos en una empresa de streaming competidora de Netflix.
Necesito entender qu√© g√©neros dominan el cat√°logo de Netflix para nuestra estrategia.

DATOS:
Tengo un DataFrame 'df' con una columna 'listed_in' que contiene los g√©neros.
Ejemplos de valores:
- 'Documentaries'
- 'International Movies, Dramas, Comedies'
- 'TV Dramas, International TV Shows'

Nota: Una pel√≠cula puede tener m√∫ltiples g√©neros separados por coma.

OBJETIVO:
1. Separar los g√©neros (split por coma)
2. Contar la frecuencia de cada g√©nero individual
3. Mostrar los 15 g√©neros m√°s populares

RESTRICCIONES:
- Usar pandas y matplotlib
- El gr√°fico debe tener t√≠tulo y etiquetas claras

FORMATO DE SALIDA:
C√≥digo Python comentado que pueda ejecutar en Google Colab
```

### 7.3 Ejecutemos este an√°lisis de g√©neros:

````python
# An√°lisis de g√©neros en Netflix

# Paso 1: Separar los g√©neros (cada t√≠tulo puede tener varios)
# Usamos explode() para crear una fila por cada g√©nero
generos = df['listed_in'].str.split(', ').explode()

# Paso 2: Contar la frecuencia de cada g√©nero
conteo_generos = generos.value_counts().head(15)

# Paso 3: Crear el gr√°fico
plt.figure(figsize=(12, 6))
conteo_generos.plot(kind='barh', color='teal', edgecolor='white')

plt.title('üé¨ Top 15 G√©neros m√°s Populares en Netflix', fontsize=14, fontweight='bold')
plt.xlabel('Cantidad de T√≠tulos')
plt.ylabel('G√©nero')

# Invertir el orden para que el m√°s popular quede arriba
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()

# Mostrar los n√∫meros exactos
print("\nüìä Conteo exacto por g√©nero:")
print(conteo_generos)
````
