#### **1. Introducci√≥n a las APIs: Las Puertas de Acceso a los Datos del Mundo**

API (Interfaz de Programaci√≥n de Aplicaciones). Una buena analog√≠a es pensar en una API como un mesero en un restaurante:

*   **Usted (el cliente/analista):** Quiere obtener algo espec√≠fico (un plato de comida/datos).
*   **La Cocina (el servidor/fuente de datos):** Tiene lo que usted necesita, pero no puede entrar directamente a la cocina a prepararlo.
*   **El Mesero (la API):** Toma su pedido (petici√≥n), lo lleva a la cocina y le trae de vuelta lo que solicit√≥ (respuesta).

En el mundo digital, las APIs permiten que diferentes aplicaciones se comuniquen entre s√≠ e intercambien datos de una manera estandarizada. Para un analista de datos, esto significa acceso a datos en tiempo real de diversas fuentes: redes sociales, mercados financieros, informaci√≥n gubernamental, datos meteorol√≥gicos y mucho m√°s.

#### **2. Selecci√≥n de APIs Gratuitas para Proyectos de An√°lisis**

| **OpenWeatherMap** | Ofrece datos meteorol√≥gicos actuales y pron√≥sticos para ciudades de todo el mundo. | S√≠ (para obtener una API Key gratuita) | Trabajar con datos num√©ricos, series de tiempo y geolocalizaci√≥n. |

| **TheCocktailDB** | Una base de datos de recetas de c√≥cteles con ingredientes, im√°genes e instrucciones. | S√≠ (para obtener una API Key gratuita) | An√°lisis de texto y datos categ√≥ricos. |

Para esta clase, vamos a utilizar **REST Countries** por su simplicidad y la riqueza de sus datos.

**Paso 2: Importar las librer√≠as necesarias**

Necesitaremos `requests` para hacer la petici√≥n a la API y `pandas` para manejar los datos.

```python
import requests
import pandas as pd
```

**Paso 3: Realizar la petici√≥n a la API**

Vamos a pedirle a la API de REST Countries toda la informaci√≥n que tiene sobre todos los pa√≠ses.

```python
# La URL del 'endpoint' que nos dar√° la informaci√≥n de todos los pa√≠ses
url = "https://restcountries.com/v3.1/all?fields=name,flags,capital,currencies,languages,population,status,region"

# Utilizamos requests.get() para hacer la petici√≥n
response = requests.get(url)

# Es una buena pr√°ctica verificar el estado de la respuesta
# Un c√≥digo 200 significa que todo sali√≥ bien.
print(f"Status Code: {response.status_code}")
```

**Paso 4: Convertir la respuesta a formato JSON**

La mayor√≠a de las APIs modernas devuelven los datos en formato JSON (JavaScript Object Notation), que es muy f√°cil de trabajar en Python.

```python
# Convertimos la respuesta en un objeto JSON (una lista de diccionarios en este caso)
data = response.json()

# Veamos la informaci√≥n de los primeros dos pa√≠ses para entender la estructura
print(data[0:2])
```

**Paso 5: Cargar los datos en un DataFrame de Pandas**

¬°Aqu√≠ es donde comienza la magia del an√°lisis de datos! Un DataFrame nos proporciona una estructura tabular con la que es mucho m√°s sencillo trabajar.

```python
# Creamos el DataFrame directamente desde la lista de diccionarios
df_paises = pd.DataFrame(data)

# Mostramos las primeras 5 filas del DataFrame
df_paises.head()
```

#### **4. Exploraci√≥n y An√°lisis de Datos (EDA)**

Una vez que los datos est√°n en un DataFrame, podemos empezar a hacer preguntas.

**Pregunta 1: ¬øCu√°les son las columnas que tenemos y qu√© tipo de datos contienen?**



```python
# Obtenemos informaci√≥n general del DataFrame
df_paises.info()
```

*Observaci√≥n:* Notar√°n que algunas columnas contienen objetos (listas o diccionarios), como 'name', 'currencies', 'languages'. ¬°Esto es interesante y requerir√° un poco de pre-procesamiento para an√°lisis m√°s profundos!



**Pregunta 2: ¬øCu√°les son los 10 pa√≠ses m√°s poblados del mundo?**

```python
# Ordenamos el DataFrame por la columna 'population' de mayor a menor y tomamos los 10 primeros
top_10_poblacion = df_paises.sort_values(by='population', ascending=False).head(32)

# Mostramos el resultado
print(top_10_poblacion[['name', 'population']])
```

*Observaci√≥n:* La columna 'name' es un diccionario. Para mostrar solo el nombre com√∫n, tendr√≠amos que acceder a √©l. Hag√°moslo m√°s limpio.

**Limpiando la columna de nombres:**

```python
# Extraemos el nombre com√∫n de la columna 'name' que es un diccionario
df_paises['common_name'] = df_paises['name'].apply(lambda x: x.get('common'))

# Ahora, repitamos la pregunta con el nombre limpio
top_10_poblacion = df_paises.sort_values(by='population', ascending=False).head(10)
print(top_10_poblacion[['common_name', 'population', 'region']])
```

**Pregunta 3: ¬øCu√°ntos pa√≠ses hay por regi√≥n (continente)?**

```python
# Contamos los valores de la columna 'region'
conteo_por_region = df_paises['region'].value_counts()

print(conteo_por_region)
```

**Visualizaci√≥n de los resultados**

Una parte crucial del an√°lisis es la visualizaci√≥n. Usemos `matplotlib` para visualizar el conteo por regi√≥n.


```python
import matplotlib.pyplot as plt

# Creamos un gr√°fico de barras
conteo_por_region.plot(kind='bar', figsize=(10, 6))
plt.title('N√∫mero de Pa√≠ses por Regi√≥n')
plt.xlabel('Regi√≥n')
plt.ylabel('Cantidad de Pa√≠ses')
plt.xticks(rotation=45)
plt.show()
```
