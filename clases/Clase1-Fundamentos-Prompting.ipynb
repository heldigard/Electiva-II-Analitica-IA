## ğŸ¯ Parte 2.5: Business Question Canvas

### Â¿QuÃ© es el Business Question Canvas?

Antes de pedirle a la IA que analice datos, debes tener **claridad sobre quÃ© quieres saber**. El **Business Question Canvas** es una plantilla simple para transformar preguntas vagas en preguntas de negocio estructuradas.

> ğŸ’¡ **AnalogÃ­a:** Antes de construir una casa, necesitas planos. Antes de analizar datos, necesitas un Business Question Canvas.

### ğŸ“‹ Los 6 Elementos de una Pregunta de Negocio

| Elemento | DescripciÃ³n | Ejemplo |
|----------|-------------|---------|
| **1. Problema** | Â¿QuÃ© problema quieres resolver? | "Las ventas cayeron 15% el Ãºltimo trimestre" |
| **2. Pregunta** | Â¿QuÃ© quieres saber exactamente? | "Â¿QuÃ© categorÃ­a de productos tiene menor rentabilidad?" |
| **3. MÃ©trica** | Â¿QuÃ© nÃºmero medirÃ¡ el Ã©xito? | Margen % = (Precio - Costo) / Precio Ã— 100 |
| **4. Audiencia** | Â¿QuiÃ©n tomarÃ¡ decisiones? | DueÃ±o de PyME, Director de Producto |
| **5. DecisiÃ³n** | Â¿QuÃ© acciÃ³n se tomarÃ¡? | Ajustar precios, eliminar productos |
| **6. Riesgo** | Â¿QuÃ© pasa si te equivocas? | PÃ©rdida de revenue, insatisfacciÃ³n de clientes |

### âœ… Ejemplo: Pregunta Mal formulada vs Bien formulada

#### âŒ Pregunta MAL formulada:
```

```
**Problemas:** No hay contexto, no hay objetivo claro, no hay decisiÃ³n asociada.

---

#### âœ… Pregunta BIEN formulada (con Canvas):

**1. Problema:** Los mÃ¡rgenes de ganancia han caÃ­do del 18% al 12% en el Ãºltimo aÃ±o. No sabemos quÃ© productos son mÃ¡s rentables.

**2. Pregunta:** Â¿CuÃ¡les son los 5 productos con menor margen de ganancia y deberÃ­amos considerar aumentar precio o discontinuar?

**3. MÃ©trica:** Margen % = (Precio - Costo) / Precio Ã— 100

**4. Audiencia:** DueÃ±o de la PyME y Director Comercial

**5. DecisiÃ³n:** Discontinuar productos con margen < 10% y bajo volumen; Aumentar precio 10-15% en productos con margen 10-13%

**6. Riesgo:** Si eliminamos los productos equivocados, podemos perder clientes y reducir revenue total.

## ğŸ” Parte 2.6: AI Reliability Checklist

â±ï¸ *Tiempo estimado: 10 minutos*

### La Regla de Oro del AnÃ¡lisis con IA

> **"ConfÃ­a, pero verifica"**
>
> La IA puede alucinar: inventar datos, interpretar mal cÃ³digo, o hacer cÃ¡lculos incorrectos.
> **TÃº eres el piloto**, la IA es tu copiloto. Siempre verifica lo que genera.

### âœ… The 3-Check Method

Para cualquier anÃ¡lisis importante, haz **3 verificaciones**:

#### Check 1: Summary Statistics
```python
# Antes de aceptar un resultado, pide ver:
df.describe()  # EstadÃ­sticas resumen
df.info()      # Estructura de datos
```
**QuÃ© verificar:**
- [ ] Los totales tienen sentido (no demaciado altos/bajos)
- [ ] No hay valores imposibles (edad: 200, satisfacciÃ³n: 7/5)
- [ ] Los tipos de datos son correctos

#### Check 2: Sample Rows
```python
# Revisa las primeras filas del resultado:
df.head(10)  # Primeras 10 filas
```
**QuÃ© verificar:**
- [ ] Los valores se ven correctos visualmente
- [ ] No hay NaN o valores extraÃ±os (inf, null)
- [ ] Los cÃ¡lculos por fila tienen sentido

#### Check 3: Triangulation
```python
# Calcula el mismo resultado de otra forma:
# MÃ©todo 1
result1 = df.groupby('categoria')['ventas'].sum()

# MÃ©todo 2 (diferente)
result2 = df.pivot_table(index='categoria', values='ventas', aggfunc='sum')

# Deben dar el mismo resultado
assert result1.equals(result2)  # Si no, investiga por quÃ©
```
**QuÃ© verificar:**
- [ ] Diferentes mÃ©todos dan el mismo resultado
- [ ] Si hay diferencia, la IA la explica lÃ³gicamente
- [ ] No hay contradicciones en las respuestas

### ğŸš¨ Signos de AlucinaciÃ³n

SeÃ±ales de que la IA podrÃ­a estar inventando resultados:

ğŸš¨ **NÃºmeros demasiado redondos**
- "El promedio es exactamente 50.0%"
- "Tienen 1,000 clientes exactamente"

ğŸš¨ **Resultados demasiado perfectos**
- "Todos los productos tienen el mismo margen"

ğŸš¨ **Falta de transparencia**
- No muestra el cÃ³digo que usÃ³
- No puede explicar cÃ³mo llegÃ³ al resultado

ğŸš¨ **Contradicciones**
- Dice "El promedio es 4.2/5" pero el histograma muestra mayorÃ­a de 1-2

### 5.1 Identificar valores faltantes

Los valores faltantes (NaN) son muy comunes en datasets reales. Es importante identificarlos **antes** de hacer cualquier anÃ¡lisis serio.

> ğŸ’¡ **Â¿Por quÃ© importan los valores faltantes?** Pueden distorsionar promedios, romper grÃ¡ficos, o hacer que tus conclusiones sean incorrectas.

````python
# Contar valores faltantes por columna
valores_faltantes = df.isnull().sum()

# Mostrar solo las columnas con valores faltantes
print("ğŸ“Š Columnas con valores faltantes:")
print("=" * 40)
print(valores_faltantes[valores_faltantes > 0].sort_values(ascending=False))
````

### 5.2 Visualizar valores faltantes

````python
# Crear grÃ¡fico de valores faltantes
plt.figure(figsize=(10, 5))

# Calcular porcentaje de valores faltantes
porcentaje_faltantes = (df.isnull().sum() / len(df) * 100).sort_values(ascending=True)

# Crear grÃ¡fico de barras horizontales
porcentaje_faltantes.plot(kind='barh', color='coral')

plt.title('ğŸ“Š Porcentaje de Valores Faltantes por Columna', fontsize=14, fontweight='bold')
plt.xlabel('Porcentaje (%)')
plt.ylabel('Columna')

# Agregar lÃ­nea vertical en 0%
plt.axvline(x=0, color='black', linewidth=0.5)

plt.tight_layout()
plt.show()
````

## ğŸ“ˆ Parte 6: AnÃ¡lisis BÃ¡sico con Prompting

### 6.1 AnÃ¡lisis por AÃ±o de Lanzamiento

Vamos a analizar cÃ³mo se distribuyen los tÃ­tulos por aÃ±o de lanzamiento.

````python
# Contar tÃ­tulos por aÃ±o de lanzamiento
titulos_por_aÃ±o = df['release_year'].value_counts().sort_index()

# Filtrar Ãºltimos 20 aÃ±os para mejor visualizaciÃ³n
titulos_recientes = titulos_por_aÃ±o[titulos_por_aÃ±o.index >= 2005]

# Crear grÃ¡fico
plt.figure(figsize=(12, 5))
titulos_recientes.plot(kind='bar', color='steelblue', edgecolor='white')

plt.title('ğŸ“Š TÃ­tulos de Netflix por AÃ±o de Lanzamiento (2005-2021)', fontsize=14, fontweight='bold')
plt.xlabel('AÃ±o de Lanzamiento')
plt.ylabel('Cantidad de TÃ­tulos')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
````

### 6.2 ComparaciÃ³n: PelÃ­culas vs Series por AÃ±o

````python
# Agrupar por aÃ±o y tipo
por_aÃ±o_tipo = df[df['release_year'] >= 2005].groupby(['release_year', 'type']).size().unstack(fill_value=0)

# Crear grÃ¡fico
fig, ax = plt.subplots(figsize=(12, 5))

por_aÃ±o_tipo.plot(kind='bar', ax=ax, width=0.8)

plt.title('ğŸ“Š PelÃ­culas vs Series por AÃ±o de Lanzamiento', fontsize=14, fontweight='bold')
plt.xlabel('AÃ±o de Lanzamiento')
plt.ylabel('Cantidad de TÃ­tulos')
plt.legend(title='Tipo')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
````

## ğŸ¯ Parte 7: TÃ©cnicas Avanzadas de Prompting

### 7.1 El Prompt de "Contexto Completo"

Cuando necesites hacer anÃ¡lisis mÃ¡s complejos, usa este formato:

```
CONTEXTO:
[Describe la situaciÃ³n y los datos]

DATOS:
[Muestra ejemplos especÃ­ficos de tus datos]

[QuÃ© quieres lograr]

[Cualquier limitaciÃ³n o preferencia]

[CÃ³mo quieres el cÃ³digo o resultado]
```

### 7.2 Ejemplo de Prompt Completo:

```
CONTEXTO:
Soy analista de datos en una empresa de streaming competidora de Netflix.
Necesito entender quÃ© gÃ©neros dominan el catÃ¡logo de Netflix para nuestra estrategia.

DATOS:
Tengo un DataFrame 'df' con una columna 'listed_in' que contiene los gÃ©neros.
Ejemplos de valores:
- 'Documentaries'
- 'International Movies, Dramas, Comedies'
- 'TV Dramas, International TV Shows'

Nota: Una pelÃ­cula puede tener mÃºltiples gÃ©neros separados por coma.

OBJETIVO:
1. Separar los gÃ©neros (split por coma)
2. Contar la frecuencia de cada gÃ©nero individual
3. Mostrar los 15 gÃ©neros mÃ¡s populares

RESTRICCIONES:
- Usar pandas y matplotlib
- El grÃ¡fico debe tener tÃ­tulo y etiquetas claras

FORMATO DE SALIDA:
CÃ³digo Python comentado que pueda ejecutar en Google Colab
```

### 7.3 Ejecutemos este anÃ¡lisis de gÃ©neros:

````python
# AnÃ¡lisis de gÃ©neros en Netflix

# Paso 1: Separar los gÃ©neros (cada tÃ­tulo puede tener varios)
# Usamos explode() para crear una fila por cada gÃ©nero
generos = df['listed_in'].str.split(', ').explode()

# Paso 2: Contar la frecuencia de cada gÃ©nero
conteo_generos = generos.value_counts().head(15)

# Paso 3: Crear el grÃ¡fico
plt.figure(figsize=(12, 6))
conteo_generos.plot(kind='barh', color='teal', edgecolor='white')

plt.title('ğŸ¬ Top 15 GÃ©neros mÃ¡s Populares en Netflix', fontsize=14, fontweight='bold')
plt.xlabel('Cantidad de TÃ­tulos')
plt.ylabel('GÃ©nero')

# Invertir el orden para que el mÃ¡s popular quede arriba
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()

# Mostrar los nÃºmeros exactos
print("\nğŸ“Š Conteo exacto por gÃ©nero:")
print(conteo_generos)
````
